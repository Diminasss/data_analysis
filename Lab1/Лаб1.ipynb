Были найдены дубликаты. Удалим их и обновим индексирование.

```python
visits = visits.drop_duplicates().reset_index(drop=True)
print(visits.shape[0])
visits[visits.duplicated()]
```

Все явные дубликаты были удалены. Стоит заметить, что были удалены обе строки, а не одна из них, так как количество строк уменьшилось на 2.

Приступим к удалению неявных дубликатов с помощью переименования. Переименуем USA в United_States, IPHONE в iPhone и MAC в Mac.

```python
visits["Region"] = visits['Region'].replace("USA", "United States")
print(visits['Region'].unique())


visits['Device'] = visits['Device'].replace("IPHONE", "iPhone")
visits['Device'] = visits['Device'].replace("MAC", "Mac")
print(visits['Device'].unique())

visits[visits.duplicated()]
```

Все явные и неявные дубликаты были удалены. Теперь будем изменять типы данных. Первым делом изменим User_Id на int64 из строчного объекта. Для этого сначала потребуется заменить запятую на точку.

```python
if visits['User_Id'].dtype == 'object':
    visits['User_Id'] = visits['User_Id'].str.replace(',', '.')

visits['User_Id'] = pd.to_numeric(visits['User_Id'],errors='raise')
visits['User_Id'] = visits['User_Id'].astype('int64')
print(visits['User_Id'].dtype)
```

Столбцы Session_Start и Session_End были успешно конвертированы в datetime64. Все столбцы были приведены к нужным значениям. Приступим далее к созданию сводных таблиц и группировок.

# Дополнительное задание на 5

1. Сделать отдельный столбец с длительностью сессии в минутах.

2. Присвоить длительности сессий 3 категории: короткая, средняя, длинная.

3. Выбрать те записи, которые больше среднего количества длительности.

4. Создать сводную таблицу категория длительности + устройство + уникальное количество пользователей.

5. Описать разницу между функциями count, nunic и другими функциями. 




Начнём с создания дополнительного столбца с длительностью сессий в минутах.

```python
visits["Duration_Minutes"] = (visits['Session_End'] - visits['Session_Start']).dt.total_seconds() / 60
visits
```

В таблице появился столбец Duration_Minutes, в котором записана длительность каждой сессии в минутах. Используется тип данных float64, так как длительность сессии может быть дробным числом в минутах.

Для определения категорий длительности сессии узнаем максимальное и минимальное время сессии. Далее поделим на 3 равных интервала.

```python
min_val = visits['Duration_Minutes'].min()
max_val = visits['Duration_Minutes'].max()

interval_1 = [int(min_val), round(float((max_val - min_val) / 3))]
interval_2 = [int(interval_1[1]), round(float(((max_val - min_val) / 3) * 2))]
interval_3 = [int(interval_2[1]), int(max_val)]

print(f"Минимальное: {min_val}")
print(f"Максимальное: {max_val}\n")
print(f"Интервал 1: [{interval_1[0]}, {interval_1[1]})")
print(f"Интервал 2: [{interval_2[0]}, {interval_2[1]})")
print("Интервал 3:", interval_3)
```

Интервал 1 - короткая сессия, интервал 2 - средняя сессия и интервал 3 - длинная сессия. Создадаим новый столбец, в котором присвоим категории длинам интервалов. Для этого воспользуемся функцией cut.

```python
bins = [int(min_val), round(float((max_val - min_val) / 3)), round(float(((max_val - min_val) / 3) * 2)), int(max_val) + 1]
labels = ['Interval 1', 'Interval 2', 'Interval 3']

visits['Interval'] = pd.cut(visits['Duration_Minutes'], bins=bins, labels=labels, right=False)
visits.sort_values(by='Duration_Minutes', ascending=False)
```

Сессии были разделены виды:

Interval 1 - Короткая

Interval 2 - Средняя

Interval 3 - Длинная

Теперь выберем записи, которые больше среднего количества длительности.

```python
medium_value = visits['Duration_Minutes'].mean() # Средняя продолжительность сессии
print("Среднее значение:", medium_value)

visits_above_mean = visits[visits['Duration_Minutes'] > medium_value].reset_index(drop=True)
visits_above_mean
```

В таблице visits_above_mean содержатся строки, в которых продолжительность сессии выше среднего. Создадим сводную таблицу по заданию "длительности + устройство + уникальное количество пользователей"

```python
additional_pivot_table: pd.DataFrame = visits.pivot_table(index=['Interval'], columns=['Device'], values='User_Id', aggfunc=['nunique'], observed=False).reset_index()
print(additional_pivot_table)
```

По текущей группировке большинство сессий является короткими, средние сессии в меньшинстве, а длинные и вовсе имеются только в одном экземпляре, принадлежит она пользователю iPhone.

Далее будет выполнено задание "Описать разницу между функциями count, nunic и другими функциями." 

1. count() считает количество непустых (ненулевых и не NaN) значений в столбце или строке.
2. nunique() считает количество уникальных значений в столбце (считая только ненулевые и не NaN значения).
3. unique() возвращает массив всех уникальных значений в столбце в том порядке, в котором они встречаются. В отличие от nunique(), которая возвращает количество уникальных значений, unique() возвращает сами уникальные элементы.
4. mean() вычисляет среднее арифметическое для числовых значений.
5. min() и max() определяют минимальное и максимальное значение в столбце.
6. std() вычисляет стандартное отклонение значений.
7. median() возвращает медиану, то есть центральное значение отсортированного ряда.
8. mode() возвращает самое часто встречающееся значение в столбце (моду).
9. describe() Возвращает общие статистические данные по столбцу или DataFrame (включает count, mean, std, min, 25%, 50%, 75%, и max).
10. quantile() возвращает значение на указанном квантиле (процентиле).
11. sum() суммирует все значения в столбце. Для строк или логических значений может работать иначе (например, считать количество True значений).

# Второе дополнительное задание на 5
1. Выбрать топ 2 канала по количеству записей
2. Выбрать топ 2 устройства по средней длительности сессии
3. Для них построить сводную таблицу канал устойство-средняя длительность сессии

```python
# 1. Топ-2 канала по количеству записей
# Выбор Топ-2 канала по количеству записей
top_2_channels = visits['Channel'].value_counts().nlargest(2).index
print(f"Топ 2 канала по количеству записей: {list(top_2_channels)}")

# Создание нового датафрейма, содержащего только данные об этих каналах
visits_top_channels = visits[visits['Channel'].isin(top_2_channels)].reset_index(drop=True)
print(f"Каналы в новом датафрейме: {visits_top_channels['Channel'].unique()}")
visits_top_channels
```

Был собран датафрейм, в котором содержатся сведения только о тех записях, которые относятся к каналам organic и Faceboom. Приступим ко второму пункту.

```python
# 2. Топ-2 устройства по средней длительности сессии
# Выбор Топ-2 устройства по средней длительности сессии
top_2_devices = visits.groupby("Device")['Duration_Minutes'].mean().nlargest(2).index
print(f"Топ 2 устройства по средней длительности сессии: {list(top_2_devices)}")

# Создание нового датафрейма, содержащего данные об этих устройствах
visits_top_devices = visits[visits['Device'].isin(top_2_devices)].reset_index(drop=True)
print(f"Каналы в новом датафрейме: {visits_top_devices['Device'].unique()}")
visits_top_devices
```

Был создан датафрейм, который содержит только те записи, которые имеют отношение к Android и PC. Построим для них сводную таблицу канал устойство-средняя длительность сессии

```python
# У меня есть желание сделать третье задание 2 способами: слиянием двух полученных таблиц и фильтрацией главного датафрейма.
# Начнём со слияния датафремов.
visits_filtred = pd.merge(visits_top_channels, visits_top_devices, on=["User_Id", "Region", "Device", "Channel", "Session_Start", "Session_End", "Duration_Minutes", "Interval"], how='inner')
print(f"Фильтрованный датафрейм содержит {visits_filtred['Channel'].unique()} и {visits_filtred['Device'].unique()}.")
# Создание сводной таблицы
visits_filtred_pivot_table = visits_filtred.pivot_table(index="Device", columns="Channel", values="Duration_Minutes", aggfunc="mean").reset_index()
visits_filtred_pivot_table
```

Третье задание было выполнено с помощью одного метода, теперь сделаем его с помощью второго

```python
# Выполним третье задание с помощью фильтрации основного датафрейма
visits_filtred_2 = visits[(visits['Channel'].isin(top_2_channels)) & (visits['Device'].isin(top_2_devices))].reset_index(drop=True)
print(f"Фильтрованный датафрейм содержит {visits_filtred_2['Channel'].unique()} и {visits_filtred_2['Device'].unique()}.")
visits_filtred_pivot_table_2 = visits_filtred_2.pivot_table(index="Device", columns="Channel", values="Duration_Minutes", aggfunc="mean").reset_index()
visits_filtred_pivot_table_2
```

По данной таблице можно сказать, что пользователи Android имеют большую продолжительность с канала organic, а пользователи PC с канала FaceBoom. Это может говорить о том, что каналы лучше подходят для разной аудитории.
